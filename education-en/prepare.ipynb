## Carrega o volabulário eric no Annif

```python
import subprocess

annif_vocab = 'eric'
out = subprocess.run('annif load-vocab ' + annif_vocab + ' --language en ./prepared-files/vocabulary/eric_thesaurus-2025.tsv', stdout=subprocess.PIPE, shell=True)
print(out.stdout)
```

## Realiza os treinamentos definidos no arquivo de projetos

```python
annif_projects = [
    'tfidf-eric-en',
    'fasttext-eric-en',
    'omikuji-parabel-eric-en',
    'mllm-eric-en',
    'ensemble-eric-en'
]

for p in annif_projects:
    out = subprocess.run('annif train ' + p + ' ./prepared-files/corpus-training/', stdout=subprocess.PIPE, shell=True)
    print(out.stdout)
```
