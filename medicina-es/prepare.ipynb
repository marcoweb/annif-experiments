## Carrega o volabulário bireme no Annif

```python
import subprocess

annif_vocab = 'bireme'
out = subprocess.run('annif load-vocab ' + annif_vocab + ' --language es ./prepared-files/vocabulary/bireme_decs_spa2022.tsv', stdout=subprocess.PIPE, shell=True)
print(out.stdout)
```

## Realiza os treinamentos definidos no arquivo de projetos

```python
annif_projects = [
    'tfidf-bireme-es',
    'fasttext-bireme-es',
    'omikuji-parabel-bireme-es',
    'mllm-bireme-es',
    'ensemble-bireme-es'
]

for p in annif_projects:
    out = subprocess.run('annif train ' + p + ' ./prepared-files/corpus-training/', stdout=subprocess.PIPE, shell=True)
    print(out.stdout)
```
